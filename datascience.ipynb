```python
import matplotlib.pyplot as plt
```

```python
#using Line Plot
plt.plot(df['Exam_id'],df['Mark'],linestyle='--',color='b',linewidth=2,marker='o',markersize=10,markerfacecolor='red',markeredgecolor='black',markeredgewidth=2,alpha=0.8)
plt.title('Customized Line Plot')
plt.xlabel('Exam_id')
plt.ylabel('Mark')
plt.grid(True,color='k')
plt.show()
```

```python
#using Horizontal Bar chart
plt.barh(df['Exam_id'],df['Mark'],linewidth=0.2,edgecolor='black',alpha=0.4)
plt.title('Bar chart')
plt.xlabel('Exam_id')
plt.ylabel('Mark')
plt.show()
```

```python
#using pie chart
color=['blue','pink','green','purple','yellow','red','orange']
plt.pie(df['Mark'],labels=df['Exam_id'],autopct='%1.1f%%',colors=color)
plt.title('Pie Chart')
plt.show()
```

```python
#using Histogarm
plt.hist(df['Mark'],bins=20,color='blue',edgecolor='white')
plt.title('Histogram')
plt.xlabel('Exam_id')
plt.ylabel('Mark')
plt.show()
```
